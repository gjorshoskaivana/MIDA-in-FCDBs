```python
category_ids = np.unique(data['food_category_id'])
```

```python
data_temp = data.copy()
data_temp
```

```python
del data_temp['fdc_id']
del data_temp['food_category_id']
```

```python
data_temp = pd.DataFrame(data_temp, columns=columns[2:])
```

```python
data_temp
```

```python
food_ids = data['fdc_id']
food_ids
```

```python
categories_ids = data['food_category_id']
```

```python
data_temp.insert(loc=0, column='fdc_id', value=food_ids)
```

```python
data_temp.insert(loc=1, column='food_category_id', value=categories_ids)
```

```python
data_temp
```

```python
categories_mean = pd.DataFrame(columns = columns)
categories_median = pd.DataFrame(columns = columns)

for category in category_ids:
  cat = data_temp.loc[data_temp['food_category_id'] == category]
  mean = pd.Series(cat.mean(skipna=True), index=columns)
  median = pd.Series(cat.median(skipna=True), index=columns)
  categories_mean = categories_mean.append(mean, ignore_index = True)
  categories_median = categories_median.append(median, ignore_index = True)
```

```python
del categories_mean['fdc_id']
```

```python
del categories_median['fdc_id']
```

```python
foods_and_cat = data_temp[['fdc_id', 'food_category_id']]
```

```python
foods_and_cat
```

```python
df_mean = pd.merge(foods_and_cat, categories_mean, on='food_category_id')
df_mean
```

```python
df_mean.to_csv('datasets/average datasets/traditional methods/food_mean.csv')
```

```python
df_median = pd.merge(foods_and_cat, categories_median, on='food_category_id')
df_median
```
