```python
import numpy as np
import pandas as pd
```

```python
food_nutrient = pd.read_csv('datasets/original data/food_nutrient.csv')
food_nutrient
```

```python
food_nutrient.isna().sum()
```

```python
food_ids = np.unique(food_nutrient['fdc_id'])
len(food_ids)
```

```python
nutrient_ids = np.unique(food_nutrient['nutrient_id'])
len(nutrient_ids)
```

```python
data = []
for food in food_ids:
  pom = food_nutrient[food_nutrient['fdc_id'] == food]
  row = []
  for nutrient in nutrient_ids:
    pom2 = pom[pom['nutrient_id'] == nutrient]
    if len(pom2) == 0:
      row.append(0)
    else:
      row.append(pom2['amount'].tolist()[0])
  data.append(row)
data
```

```python
dataset = pd.DataFrame(data)
dataset.columns = nutrient_ids
dataset.index = food_ids
dataset
```

```python
dataset.to_csv('datasets/average datasets/byColumns/average_dataset_198.csv')
```
